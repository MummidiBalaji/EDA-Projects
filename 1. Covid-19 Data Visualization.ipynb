### Top 10 states having high total cases

```python
high_cases = covid.sort_values(by='Total Cases',ascending=False).head(10)
xpos = np.arange(len(high_cases["State/UTs"]))
fig = plt.subplots(figsize=(15,6))
plt.xticks(xpos,high_cases["State/UTs"])
plt.bar(high_cases["State/UTs"],high_cases["Total Cases"],color = 'red',width=0.5)
plt.xlabel("Name of the State")
plt.ylabel("Total Cases")
plt.title("Top 10 states by Total cases")
plt.show()
```

### Top 5 states having least total cases

```python
least_cases = covid.sort_values(by='Total Cases',ascending=True).head()
ypos = np.arange(len(least_cases["State/UTs"]))
fig = plt.subplots(figsize=(20,6))
plt.xlabel("Name of the state")
plt.ylabel("Total Cases")
plt.title("Top 5 states having least cases")
plt.xticks(ypos,least_cases["State/UTs"])
plt.bar(least_cases["State/UTs"],least_cases["Total Cases"],color = 'orange',width=0.5)
plt.show()
```
